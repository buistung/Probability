Algorithms

```python
import matplotlib.pyplot as plt

def same_birthday_probability(n):
    not_same_prob = 1.0
    for i in range(n):
        not_same_prob *= (365 - n) / 365

    return 1 - not_same_prob


```

Test

```python
people_range = range(1,101)
probabilities = [same_birthday_probability(n) for n in people_range]
```

Special cases

```python
print('Probability to have the same birthday:')
special_case  = [23,30,50,70]
for i in special_case:
    prob = same_birthday_probability(i)
    print(f'{i} people: {prob:.2%}')
```

Plotting result

```python
plt.figure(figsize=(10, 6))  
plt.plot(people_range, probabilities, 'b-')  
plt.xlabel('Number of people')  
plt.ylabel('Probability of at least one shared birthday')  
plt.title('Birthday paradox: Probability vs. Number of people')  


plt.axhline(y=0.5, color='r', linestyle='--', alpha=0.5)  


plt.axvline(x=23, color='r', linestyle='--', alpha=0.5)  

plt.show()  
```
