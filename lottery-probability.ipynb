Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Set up simulation parameters

```python
n_simulations = 100000
numbers_range = 10
ticket_length = 6
```

Simulation

```python
# array for saving winning times
matches_count = np.zeros(ticket_length + 1)

# Simulation
for _ in range(n_simulations):
    player_numbers = np.random.randint(0, numbers_range, ticket_length)

    winning_numbers = np.random.randint(0, numbers_range, ticket_length)

    matches = np.sum(player_numbers == winning_numbers)
    matches_count[matches] += 1

probabilities = matches_count / n_simulations
```

Print result

```python
print(f'Simulation result after {n_simulations:,} times:')
for i in range(ticket_length + 1):
    print(f'Probability for winning {i} numbers: {probabilities[i]:.6f} ({matches_count[i]:.0f} times)')
```

Plotting result

```python

x = list(range(ticket_length + 1))
y = probabilities

plt.figure(figsize=(10, 6))
plt.bar(x, y)

plt.xlabel('Number of Matches')
plt.ylabel('Probability')
plt.title(f'Probability Distribution over {n_simulations:,} Simulations')
plt.xticks(x)  # Show all match numbers on x-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()


```
